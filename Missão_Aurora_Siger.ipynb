####Definindo as variáveis iniciais para valores a serem atribuídos

```python
temperatura_interna = 0
temperatura_externa = 0
integridade_estrutural = 0
energia = 0
pressao_tanque = 0
modulos_criticos = 0
```

Apresentamos uma função específica que recebe a voltagem e a amperagem, fazendo o cálculo automaticamente

```python
def dimensionar_energia_missao(voltagem_sistema, capacidade_ah):
    energia_wh = voltagem_sistema * capacidade_ah
    energia_kwh = energia_wh / 1000
    energia_joules = energia_wh * 3600
    return energia_kwh, energia_joules

voltagem_sistema = 400
capacidade_ah = 25000
```

Importando a biblioteca random (simulador de sensores)

```python
import random as aleatorio
```

Colocando a biblioteca random com o código 'randint' para deixar uma randomização de números inteiros

```python
temperatura_interna = aleatorio.randint(10,30)
temperatura_externa = aleatorio.randint(-20,35)
energia = aleatorio.randint(50, 100)
pressao_tanque = aleatorio.randint(20,60)
integridade_estrutural = aleatorio.randint(0,1)
modulos_criticos = aleatorio.randint(0,1)
```

Declarando os componentes da aurora como Margem = 0 (para um cálculo futuro)

```python
temperatura_inter_margem = 0
temperatura_externa_margem = 0
energia_margem = 0
pressao_tanque_margem = 0
integridade_margem = 0
modulos_criticos_margem = 0
```

Mínimo necessário para os requisitos da aurora

```python
temperatura_intermin = 10
temperatura_externamin = -10
energiamin = 60
pressao_tanquemin = 40
integridade_estruturalmin = 1
modulos_criticosmin =  1
```

Máximo que os requisitos podem chegar na aurora

```python
temperatura_intermax = 30
temperatura_externamax = 35
energiamax = 100
pressao_tanquemax = 60
integridade_max = 1
modulos_criticosmax =  1
```

Calculando a margem de diferença de quanto falta pro mínimo ou quanto passou do mínimo/máximo

```python
temperatura_inter_margem = temperatura_interna - temperatura_intermin
temperatura_externa_margem = temperatura_externa - temperatura_externamin
energia_margem = energia - energiamin
pressao_tanque_margem = pressao_tanque - pressao_tanquemin
integridade_margem = integridade_estrutural - integridade_estruturalmin
modulos_criticos_margem = modulos_criticos - modulos_criticosmin
```

Cálculo de autonomia e telemetria energética

```python
capacidade_total_kwh, energia_total_joules = dimensionar_energia_missao(voltagem_sistema, capacidade_ah)

consumo_decolagem_kwh = 1500
perdas_energeticas_kwh = 500

carga_atual_kwh = capacidade_total_kwh * (energia / 100)
carga_minima_kwh = 6000
carga_margem_kwh = carga_atual_kwh - carga_minima_kwh
autonomia_inicial = (carga_atual_kwh - consumo_decolagem_kwh) - perdas_energeticas_kwh
autonomia_minima = 0
autonomia_margem = autonomia_inicial - autonomia_minima
```

Mostrando os parâmetros que a nave tem no momento de forma aleatória!

```python
print("verificando status da nave...\n")

print(f"--- STATUS DO HARDWARE: {voltagem_sistema}V | {capacidade_ah}Ah | ENERGIA TOTAL: {energia_total_joules:,.0f} J ---\n")

print("\n------------------ TELEMETRIA E ANÁLISE DO AURORA ------------------\n")

print("| Parâmetro                         | Status Atual | Status Mínimo | Margem |")
print("|-----------------------------------|--------------|---------------|--------|")

print(f"| Temperatura interna da nave       | {temperatura_interna:^12} | {'10°C':^13} | {temperatura_inter_margem:^6} |")
print(f"| Temperatura externa da nave       | {temperatura_externa:^12} | {'-10°C':^13} | {temperatura_externa_margem:^6} |")
print(f"| Energia total da nave             | {energia:^12} | {'60%':^13} | {energia_margem:^6} |")
print(f"| Pressão do tanque da nave         | {pressao_tanque:^12} | {'40':^13} | {pressao_tanque_margem:^6} |")
print(f"| Integridade estrutural da nave    | {integridade_estrutural:^12} | {'1':^13} | {integridade_margem:^6} |")
print(f"| Módulos críticos da nave          | {modulos_criticos:^12} | {'1':^13} | {modulos_criticos_margem:^6} |")
print(f"| Energia total (Porcentagem)       | {energia:^12} | {'60%':^13} | {energia_margem:^6} |")
print(f"| Carga atual em kWh                | {carga_atual_kwh:^12.0f} | {'6000':^13} | {carga_margem_kwh:^6.0f} |")
print(f"| Autonomia estimada (Pós-Voo)      | {autonomia_inicial:^12.0f} | {'0':^13} | {autonomia_margem:^6.0f} |")

print("\n---------------------------------------------------------------------\n")
```

Definindo o status como OK ou N/G (Not good) com lógica de faixa segura

```python
status_temp_int = "OK" if temperatura_intermin <= temperatura_interna <= temperatura_intermax else "N/G"
status_temp_ext = "OK" if temperatura_externamin <= temperatura_externa <= temperatura_externamax else "N/G"
status_energia = "OK" if energiamin <= energia <= energiamax else "N/G"
status_pressao = "OK" if pressao_tanquemin <= pressao_tanque <= pressao_tanquemax else "N/G"
status_integridade = "OK" if integridade_estrutural == integridade_estruturalmin else "N/G"
status_modulos = "OK" if modulos_criticos == modulos_criticosmin else "N/G"
status_autonomia = "OK" if autonomia_inicial > 0 else "N/G"
```

Exibindo se os status da Aurora estão OK ou N/G

```python
print(f"""
Integridade estrutural da Aurora = {status_integridade}
Módulos Críticos do foguete = {status_modulos}
Temperatura interna = {status_temp_int}
Temperatura externa = {status_temp_ext}
Pressão do tanque = {status_pressao}
Energia = {status_energia}
Autonomia de Voo (kWh) = {status_autonomia}
""")
```

Definindo se a Aurora pode decolar a partir dos Status OK e N/G

```python
decisao_final = [status_integridade, status_modulos, status_temp_int, status_temp_ext, status_pressao, status_energia, status_autonomia]
if "N/G" in decisao_final:
    print("""               -------------------------------------------
               |           DECOLAGEM ABORTADA            |
               -------------------------------------------""")

else:
    print("""               ---------------------------------------------------------
               |  PRONTO PARA DECOLAR! VAMOS BALANÇAR O FUTURO!  |
               ---------------------------------------------------------""")
```
